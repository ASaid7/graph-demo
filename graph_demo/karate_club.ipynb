```python
import networkx as nx
from networkx.algorithms.community import modularity
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.cluster import SpectralClustering, KMeans
from sklearn.manifold import SpectralEmbedding
import community as community_louvain
from sklearn.metrics import normalized_mutual_info_score as NMI
%matplotlib inline
```

# Python for Network Analysis
We'll start with getting an understanding of the APIs used to analyze the Zachary's Karate Club netwrok. Then we'll expand into larger graphs with the same methods.

```python
g = nx.karate_club_graph()
adj = nx.adjacency_matrix(g)
print(nx.info(g))
```

In networkx a Graph is an undirected graph, DiGraph is a directed graph, and Multi(Di)Graph is a graph with self loops

```python
pos=nx.spring_layout(g,seed=37)
fig = plt.figure(figsize=(6,6))
nx.draw(g,pos=pos, node_color = 'darkgreen', node_size=200,
        with_labels=True, font_color='white')
plt.axis('off')
plt.show()
```

When the graph is too large its best to just view the degree distribution

```python
degrees = list(g.degree())
x,y = zip(*degrees)
fig, axs = plt.subplots(1,2,figsize=(20,6))
axs[0].hist(y)
axs[0].set_xlabel('Degrees')
axs[0].set_ylabel('Frequency')
axs[0].set_title('Degree Distribution')
axs[1].bar(x,y)
axs[1].set_xlabel('Nodes')
axs[1].set_ylabel('Degrees')
axs[1].set_title('Bar Plot')
plt.tight_layout()
plt.show()
```

Lets use pagerank & degree centrality to find the most influential karate club members, safe assumption we should get node 0 and 33 in the top 2 for both given that its their clubs

```python
pr = nx.pagerank(g)
bc = nx.betweenness_centrality(g)

print(f'Top 5 PageRank Nodes:\n {sorted(pr.items(), key=lambda x: x[1], reverse=True)[:5]}')
print(f'Top 5 Betweenness Centrality Nodes:\n {sorted(bc.items(), key=lambda x: x[1], reverse=True)[:5]}')
```

We can use Sklearn to find the eigenvectorsz, affinity is set to precomputed we dont need sklearn to do anything to our matrix

```python
embedding = SpectralEmbedding(n_components=3, affinity='precomputed')
vectors = embedding.fit_transform(adj)
```

```python
x = np.linspace(0,len(g)-1,len(g))
fig, axs = plt.subplots(2,2, figsize=(20,12))
axs = np.ravel(axs)
for i in range(3):
    axs[i].bar(x,vectors[:,i])
    axs[i].set_xlabel('Nodes')
    axs[i].set_xticks(x)
axs[3].scatter(vectors[:,0],vectors[:,1])
axs[3].set_xlabel('2nd Eigenvector')
axs[3].set_ylabel('3rd Eigenvector')
axs[0].set_title('2nd Eigenvector')
axs[1].set_title('3rd Eigenvector')
axs[2].set_title('4th Eigenvector')
plt.tight_layout()
plt.show()

ax = plt.axes(projection='3d')
ax.view_init(30, 75)
ax.scatter3D(vectors[:,0],vectors[:,1],vectors[:,2])
plt.show()
```

Given we know that there are only two communities the second eigenvector (note sklearn doesnt give you the first eigenvector unless you specify you want it) will be enough to cluster our data. When creating a scatter plot from the second and third eigenvector we start to see 3 clusters form

```python
def community_dict(nodes, labels = None, community = None):
    
    if community:
        comm = {}
        for i, v in community.items():
            comm[v] = [i] if v not in comm.keys() else comm[v] + [i]
        return(comm)
    
    elif list(labels):
        comm = {}
        pred = {node:pred for node,pred in zip(nodes, labels)}
        for i, v in pred.items():
            comm[v] = [i] if v not in comm.keys() else comm[v] + [i]
        return(comm)
    
```

```python
labels=[g._node[i]['club'] for i in g.nodes()]
```

The rare case that you have labels for your clustering problem use normalized mutual information (NMI) to measure how well your clustering algorithm performs. The closer you are to 1 the closer you are to the truth

Using the second eigenvector only leads to the best performance try out the other 3 to see the impact

```python
km = KMeans(n_clusters = 2, n_init=10).fit(vectors[:,:1])
comm = community_dict(nodes = g.nodes(), labels = km.labels_)
print(f'Normalized Mutal Information: {NMI(labels, km.labels_)}')
print(f'Modularity: {modularity(g,comm.values(),resolution=1)}')
```

Here we apply the louvain algorithm and set our resolution to 2 to control for the size of the community and we get similar performance to spectral clustering

```python
resolution = 2
partition = community_louvain.best_partition(nx.Graph(g), resolution=resolution)
comm = community_dict(nodes = g.nodes(), community=partition)
pred = [i for i in partition.values()]
print(f'Normalized Mutal Information: {NMI(labels, pred)}')
print(f'Modularity: {modularity(g,comm.values())}')
print(f'Number of Communities : {len(comm.values())}')
```

```python
sc = ['darkgreen' if km.labels_[i]==1 else 'maroon' for i in g.nodes()]
lc = ['darkgreen' if pred[i]==0 else 'maroon' for i in g.nodes()]
tc = ['maroon' if labels[i]=='Mr. Hi' else 'darkgreen' for i in g.nodes()]
fig, axs = plt.subplots(1,3, figsize=(20,8))
nx.draw(g,pos=pos, node_color = tc, node_size=200,
        with_labels=True, font_color='white',ax=axs[0])
nx.draw(g,pos=pos, node_color = sc, node_size=200,
        with_labels=True, font_color='white', ax=axs[1])
nx.draw(g,pos=pos, node_color = lc, node_size=200,
        with_labels=True, font_color='white', ax=axs[2])
axs[0].set_title('Ground Truth')
axs[1].set_title('Spectral Clustering')
axs[2].set_title('Louvain Method')
plt.show()
```
